### Lista 2


```python
import sympy as sp

sp.init_printing()
```

```python
t = sp.symbols("t")

x = 3 * (t - sp.sin(t))
y = 2 * (1 - sp.cos(t))

p1 = sp.plot_parametric(x, y, (t, 0, 2 * sp.pi))
```

### Zadanie 1


```python
xd1 = sp.diff(x, t)
yd1 = sp.diff(y, t)

xyd1 = yd1 / xd1
xyd1
```

```python
# pochodna pierwszego stopnia wartość pierwsza
xyd1_val1 = xyd1.subs(t, sp.pi / 3)
print(xyd1_val1.evalf(6))
xyd1_val1
```

```python
# pochodna pierwszego stopnia wartość druga
xyd1_val2 = xyd1.subs(t, 4 * sp.pi / 3)
print(xyd1_val2.evalf(6))
xyd1_val2
```

```python
xyd2 = xyd1 / xd1
xyd2
```

```python
# pochodna drugiego stopnia wartość pierwsza
xyd2_val1 = xyd2.subs(t, sp.pi / 4)
print(xyd2_val1.evalf(6))
xyd2_val1
```

```python
# pochodna drugiego stopnia wartość druga
xyd2_val2 = xyd2.subs(t, 5 * sp.pi / 4)
print(xyd2_val2.evalf(6))
xyd2_val2
```

```python
xyd3 = xyd2 / xd1
xyd3
```

```python
# pochodna trzeciego stopnia wartość pierwsza
xyd3_val1 = xyd3.subs(t, sp.pi / 6)
print(xyd3_val1.evalf(6))
xyd3_val1
```

```python
# pochodna drugiego stopnia wartość druga
xyd3_val2 = xyd3.subs(t, 7 * sp.pi / 6)
print(xyd3_val2.evalf(6))
xyd3_val2
```

### Zadanie 2


```python
t0 = sp.pi / 2
x0 = x.subs(t, t0)
print(x0)
y0 = y.subs(t, t0)
print(y0)
```

```python
# symbol x jako xs
xs = sp.symbols("x")

xyd1_t0value = xyd1.subs(t, t0)

tangent = xyd1_t0value * (xs - x0) + y0
tangent
```

```python
p1.append(sp.plot(tangent, show=False)[0])
p1.show()
```

### Zadanie 3


```python
dsurface = sp.Integral(sp.Abs(y * xd1), (t, 0, 2 * sp.pi))

dsurface.n(5)
```

```python
x1 = x.subs("t", 0)
print(x1)
x2 = x.subs("t", 2 * sp.pi)
print(x2)
```

### Zadanie 4


```python
L = sp.Integral(sp.sqrt(xd1**2 + yd1**2), (t, 0, 2 * sp.pi))
print(L.n(6))
L
```

### Zadanie 5


```python
volume_integral = sp.Integral(y**2 * xd1, (t, 0, 2 * sp.pi))
volume = sp.pi * volume_integral.n(5)
print(volume.evalf(10))
volume
```

### Zadanie 6


```python
surface_integral = sp.Integral(y * sp.sqrt(xd1**2 + yd1**2), (t, 0, 2 * sp.pi))
surface = 2 * sp.pi * surface_integral.n(5)
print(surface.evalf(5))
surface
```
