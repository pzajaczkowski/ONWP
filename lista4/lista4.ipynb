### Zadanie 2
zamiana:
x=rcos(t), y=rsin(t), J=r

a)

```python
# a
plt1 = plot_implicit(Eq(x ** 2 + y ** 2, 5), line_color="blue", show=False)
plt2 = plot_implicit(x, line_color="magenta", show=False)
plt3 = plot_implicit(y, x_var=x, line_color="orange", show=False)
plt4 = plot_implicit(And(x ** 2 + y ** 2 <= 5, x >= 0, y <= 0), line_color="green", show=False)
plt1.append(plt2[0])
plt1.append(plt3[0])
plt1.append(plt4[0])
plt1.show()
```

```python
# a
f = (x ** 2 + y ** 2) * exp(-3 * (x ** 2 + y ** 2))
display(f)
fs = f.subs({x: r * cos(t), y: r * sin(t)})
display(fs)
fssim = fs.simplify()
display(fssim)
```

```python
# !!domnożyć jakobian J=r
calf = fssim * r
display(Integral(calf, r, t))
cal = Integral(calf, (r, 0, sqrt(5)), (t, 3 / 2 * pi, 2 * pi))
display(cal)
display(cal.doit().evalf())
```

b)

```python
# b
plt1 = plot_implicit(Eq(x ** 2 + y ** 2, 1), line_color="blue", show=False)
plt2 = plot_implicit(Eq(x ** 2 + y ** 2, 4), line_color="magenta", show=False)
plt3 = plot_implicit(And(x ** 2 + y ** 2 <= 5, x ** 2 + y ** 2 >= 1), line_color="green", show=False)
plt1.append(plt2[0])
plt1.append(plt3[0])
plt1.show()
```

```python
# b
f = 1 / (1 + (x ** 2 + y ** 2 + 1) ** 2)
display(f)
fs = f.subs({x: r * cos(t), y: r * sin(t)})
display(fs)
fssim = fs.simplify()
display(fssim)
```

```python
# !!domnożyć jakobian J=r
calf = fssim * r
display(Integral(calf, r, t))
cal = Integral(calf, (r, 1, 2), (t, 0, 2 * pi))
display(cal)
display(cal.doit().evalf())
```

### Zadanie 4


#### Rysujemy obszar, po którym będziemy całkować


```python
x, y = symbols("x y")
p1 = plot(-((x - 2) ** 2) + 4, (x, -5, 7), show=False, line_color="magenta")
p2 = plot(2 - x, (x, -5, 10), (y, -5, 10), show=False, line_color="red")
p3 = plot_implicit(
    And(y + (x - 2) ** 2 <= 4, x + y >= 2),
    (x, -5, 5),
    (y, -5, 5),
    show=False,
    line_color="orange",
)
p1.append(p2[0])
p1.append(p3[0])
p1
p1.show()
```

```python
f = 4 - (x - 2) ** 2
f
```

```python
g = 2 - x
g
```

```python
Eq(f, g)
```

```python
s = solve(Eq(f, g))
s
```

```python
s[0].evalf(10)
```

```python
s[1].evalf(10)
```

#### sposób 1 - całka pojedyncza z różnicy funkcji f i g

```python
pole = integrate(f - g, (x, s[0], s[1]))
pole
```

```python
pole.evalf(10)
```

### Zadanie 6


#### Rysujemy obszar, po którym będziemy całkować


```python
x, y = symbols("x y")
p1 = plot(0, (x, -10, 20), show=False, line_color="magenta")
p2 = plot_implicit(
    16 - x + 3 * y, (x, -10, 20), (y, -10, 5), show=False, line_color="blue"
)
p3 = plot_implicit(x - 0, (x, -10, 20), show=False, line_color="green")
p4 = plot_implicit(
    And(16 - x + 3 * y >= 0, x >= 0, y <= 0),
    (x, -10, 20),
    (y, -10, 5),
    show=False,
    line_color="yellow",
)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1
p1.show()
```

```python
f = 16 - x + 3 * y
f
```

```python
w = sqrt(1 + f.diff(x, 1) ** 2 + f.diff(y, 1) ** 2)
w
```

#### Pole płata


```python
Integral(w, (y, (x - 16) / 3, 0), (x, 0, 16))
```

```python
integrate(w, (y, (x - 16) / 3, 0), (x, 0, 16))
```

```python
integrate(w, (y, (x - 16) / 3, 0), (x, 0, 16)).evalf(10)
```
