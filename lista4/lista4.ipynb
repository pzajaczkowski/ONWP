### Zadanie 2

zamiana:
x=rcos(t), y=rsin(t), J=r


a)


```python
# a
plt1 = plot_implicit(Eq(x**2 + y**2, 2), line_color="blue", show=False)
plt2 = plot_implicit(Eq(x**2 + y**2, 3), line_color="magenta", show=False)
plt3 = plot_implicit(
    And(x**2 + y**2 <= 3, x**2 + y**2 >= 2), line_color="green", show=False
)
plt1.append(plt2[0])
plt1.append(plt3[0])
plt1.show()
```

```python
# a
f = 2 * (x**2 + y**2) * exp(-4 * (x**2 + y**2))
display(f)
fs = f.subs({x: r * cos(t), y: r * sin(t)})
display(fs)
fssim = fs.simplify()
display(fssim)
```

```python
# !!domnożyć jakobian J=r
calf = fssim * r
display(Integral(calf, r, t))
cal = Integral(calf, (r, sqrt(2), sqrt(3)), (t, 0, 2 * pi))
display(cal)
display(cal.doit().evalf())
```

b)


```python
# b
plt1 = plot_implicit(Eq(x**2 + y**2, 1), line_color="blue", show=False)
plt2 = plot_implicit(x, line_color="magenta", show=False)
plt3 = plot_implicit(y, x_var=x, line_color="orange", show=False)
plt4 = plot_implicit(
    And(x**2 + y**2 <= 1, x <= 0, y >= 0), line_color="green", show=False
)
plt1.append(plt2[0])
plt1.append(plt3[0])
plt1.append(plt4[0])
plt1.show()
```

```python
# b
f = 1 / (x**2 + y**2 + 1) ** 3
display(f)
fs = f.subs({x: r * cos(t), y: r * sin(t)})
display(fs)
fssim = fs.simplify()
display(fssim)
```

```python
# !!domnożyć jakobian J=r
calf = fssim * r
display(Integral(calf, r, t))
cal = Integral(calf, (r, 0, 1), (t, pi / 2, pi))
display(cal)
display(cal.doit().evalf())
```

### Zadanie 4


#### Rysujemy obszar, po którym będziemy całkować


```python
f = 4 - (x - 2) ** 2
display(f)
g = 2 - x
display(g)
s = solve(Eq(f, g))
display(s)
```

```python
p1 = plot(
    f,
    (x, -5, 10),
    xlim=[-2, 5],
    ylim=[-2, 7],
    show=False,
    line_color="green",
)
p2 = plot(g, (x, -5, 10), (y, -5, 10), show=False, line_color="red")
p3 = plot_implicit(
    And(y + (x - 2) ** 2 <= 4, x + y >= 2, x >= s[0], x <= 2),
    (x, -5, 5),
    (y, -5, 5),
    show=False,
    line_color="orange",
)
p4 = plot_implicit(
    And(y + (x - 2) ** 2 >= 4, x + y <= 2, x >= 0, x <= s[0]),
    (x, -5, 5),
    (y, -5, 5),
    show=False,
    line_color="orange",
)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.show()
```

```python
Pole1Cal = Integral(1, (y, f, g), (x, 0, s[0]))
display(Pole1Cal)
Pole1 = Pole1Cal.doit().evalf()
display(Pole1)
```

```python
Pole2Cal = Integral(1, (y, g, f), (x, s[0], 2))
# display(Pole2Cal)
Pole2 = Pole2Cal.doit().evalf()
display(Pole2)
```

```python
suma_pol = Pole1 + Pole2
suma_pol
```
