```python
import sympy
import numpy as np
import matplotlib.pyplot as plt

from sympy import *

sympy.init_printing()
```

```python
x, y = symbols("x y")
```

```python
f0x, g0x, h0x = symbols("f_0(x) g_0(x) h_0(x)");

wzorF0x = x
wzorG0x = x**2
wzorH0x = x**3

p1 = plot(  wzorF0x,
            (x,-10,-1),
            xlim=[-10,10], 
            ylim=[-20,100],
            line_color = "blue",
            title = "Trzy funkcje",
            xlabel = "oś x", 
            ylabel = "oś y",
            legend = True,
            show = False)

p2 = plot(  wzorG0x,
            (x,-5,3),
            line_color = "red",
            show = False)

p3 = plot(  wzorH0x,
            (x,4,10),
            line_color = "green",
            show = False)

p1.append(p2[0])
p1.append(p3[0])

p1.show()
```

```python
punktNieciaglosci = 5
polCzastkiElementarnej = 0.01 #oksymoron :)

p1 = plot(  (1 + x ** 2, (x, 5, 100)),
            (4, (x, punktNieciaglosci - polCzastkiElementarnej, punktNieciaglosci + polCzastkiElementarnej)), 
              # pownino być (x, 5, 5), ale chciałem, żeby ten punkt też był widoczny na wykresie
            (-3 * x + 4, (x, -100, 5)),
            xlim=[punktNieciaglosci - 6, punktNieciaglosci + 6], 
            ylim=[-20, 122],
            label='f(x)',
            line_color = "blue",
            title = "Funkcja i jej pochodna",
            xlabel = "oś x", 
            ylabel = "oś y",
            legend = True,
            show = False)

p2 = plot(  (diff(1 + x ** 2), (x, 5, 100)),
            (diff(4), (x, punktNieciaglosci - polCzastkiElementarnej, punktNieciaglosci + polCzastkiElementarnej)),
            (diff(-3 * x + 4), (x, -100, 5)),
            label="f'(x)",
            line_color = "red",
            legend = True,
            show = False)

p1.append(p2[0])
p1.append(p2[1])
p1.append(p2[2])
p1.show()
```

```python
p20 = plot(-2*x+3, (x,-7,7), 
           show=False, 
           line_color='blue', 
           xlabel='OŚ X', 
           ylabel='OŚ Y',
           title = "Trójkącik",
           label='$g(x)=-2x+3$', 
           legend=True)

p21 = plot(x, (x,-2,3), 
           show=False, 
           line_color='green', 
           label='$g(x)=x$', 
           legend=True)

p22 = plot(-x, (x,-2,3), 
           show=False, 
           line_color='magenta', 
           label='$g(x)=-x$', 
           legend=True)

p23 = plot_implicit(
    And(y < x, y > -x, y<-2*x+3), (x,-5,5), 
    show=False, 
    line_color='orange')

p20.append(p21[0]) 
p20.append(p22[0]) 
p20.append(p23[0])

p20.show()
```
