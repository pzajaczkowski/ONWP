# Lista - wizualizacje


```python
import sympy as sp
from sympy import plot

sp.init_printing()

x, y = sp.symbols("x y")
```

### Zadanie 1


```python
fun_1_1 = sp.sin(x * sp.cos(x))
fun_1_2 = x ** sp.sin(x * 10) * 4
fun_1_3 = sp.log(x**2 + x**5) - 6

p1_1 = plot(
    fun_1_1,
    (x, -10, -1),
    xlabel="OŚ X",
    ylabel="OŚ Y",
    title="Zadanie 1 - 3 funkcje jednej zmiennej",
    legend=True,
    line_color="blue",
    show=False,
    xlim=[-10, 10],
    ylim=[-10, 10],
)

p1_2 = plot(
    fun_1_2,
    (x, -5, 3),
    legend=True,
    line_color="orange",
    show=False,
)

p1_3 = plot(
    fun_1_3,
    (x, 4, 10),
    legend=True,
    line_color="green",
    show=False,
)

p1_1.append(p1_2[0])
p1_1.append(p1_3[0])
p1_1.show()
```

### Zadanie 2


```python
fun2 = sp.Piecewise(
    ((x - 5) ** 2 - 25, (x >= -5) & (x < 0)),
    (x**3 + 0.5, (x >= 0) & (x < 3)),
    ((sp.sin(3 * x) * 5 + 25), (x >= 3) & (x < 6)),
)

fun2_diff = sp.diff(fun2, x)

p2_1 = plot(
    fun2,
    xlabel="OŚ X",
    ylabel="OŚ Y",
    label="f(x)",
    title="Zadanie 2 - funkcja klamerkowa i jej pochodna",
    line_color="blue",
    show=False,
    legend=True,
    xlim=[-10, 10],
    ylim=[-20, 35],
)

p2_2 = plot(fun2_diff, line_color="red", label="f '(x)", show=False)

p2_1.append(p2_2[0])
p2_1.show()
```

### Zadanie 3


```python
fun3_1 = 4 * x + 10
fun3_2 = -4 * x + 20
fun3_3 = 0.1 * x + 2

p3_1 = plot(
    fun3_1,
    line_color="blue",
    xlabel="OŚ X",
    ylabel="OŚ Y",
    title="Zadanie 3 - zamalowany trójkąt",
    legend=True,
    xlim=[-10, 10],
    ylim=[-5, 20],
    show=False,
)

p3_2 = plot(fun3_2, line_color="magenta", show=False)

p3_3 = plot(fun3_3, line_color="orange", show=False)

p3_4 = sp.plot_implicit(
    (y < fun3_1) & (y < fun3_2) & (y > fun3_3),
    (x, sp.solve(fun3_1 - fun3_3, x)[0], sp.solve(fun3_2 - fun3_3, x)[0]),
    (y, 0, 20),
    line_color="green",
    legend=False,
    show=False,
)

p3_1.append(p3_2[0])
p3_1.append(p3_3[0])
p3_1.append(p3_4[0])
p3_1.show()
```

### Zadanie 4


```python
fun4 = sp.Eq(x * sp.sin(x) - y**2 * sp.cos(x), 2)
fun4_1 = sp.Eq(x**2 + y**2, x * 10 * sp.sin(x))

p4 = sp.plot_implicit(
    fun4_1,
    (x, -4, 4),
    (y, -4, 4),
    line_color="red",
    xlabel="OŚ X",
    ylabel="OŚ Y",
    title="Zadanie 4 - Funkcja uwikłana",
)
```

### Zadanie 5

### Zadanie 6


```python
t = sp.symbols("t")

fun6_1x = 5 * sp.sin(t - 3)
fun6_1y = 5 * sp.sin(2 * t + 4)

fun6_2x = 1.5 * t * sp.cos(0.5 * t)
fun6_2y = 0.4 * t * sp.sin(2.2 * t)

p6_1 = sp.plot_parametric(
    fun6_1x,
    fun6_1y,
    (t, -10, 10),
    line_color="blue",
    xlabel="OŚ X",
    ylabel="OŚ Y",
    title="Zadanie 6 - parametryczna nr 1",
    legend=True,
)

p6_2 = sp.plot_parametric(
    fun6_2x,
    fun6_2y,
    (t, -10, 10),
    line_color="red",
    xlabel="OŚ X",
    ylabel="OŚ Y",
    title="Zadanie 6 - parametryczna nr 2",
    legend=True,
)

p6_1.title = "Zadanie 6 - obie na raz"
p6_1.append(p6_2[0])
p6_1.show()
```

### Zadanie 7


```python
from sympy.plotting import plot3d


fun7_1 = 5 * x**2 - y**2 + 3
fun7_2 = 74 * sp.sin(0.5 * x) - 102 * sp.cos(0.7 * y)

p7_1 = plot3d(
    fun7_1,
    xlabel="OŚ X",
    ylabel="OŚ Y",
    zlabel="OŚ Z",
    title="Zadanie 7 - dwóch zmiennych nr 1",
)

p7_2 = plot3d(
    fun7_2,
    xlabel="OŚ X",
    ylabel="OŚ Y",
    zlabel="OŚ Z",
    title="Zadanie 7 - dwóch zmiennych nr 2",
)

p7_1.append(p7_2[0])
p7_1.title = "Zadanie 7 - obie funkcje"
p7_1.show()
```
