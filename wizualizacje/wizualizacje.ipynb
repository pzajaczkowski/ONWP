# Lista - wizualizacje


```python
import sympy as sp
from sympy import plot

sp.init_printing()

x, y = sp.symbols("x y")
```

### Zadanie 1


```python
fun_1_1 = sp.sin(x * sp.cos(x))
fun_1_2 = x ** sp.sin(x * 10) * 4
fun_1_3 = sp.log(x**2 + x**5) - 6

p1_1 = plot(
    fun_1_1,
    (x, -10, -1),
    xlabel="x",
    ylabel="y",
    title="Zadanie 1 - 3 funkcje jednej zmiennej",
    legend=True,
    line_color="blue",
    show=False,
    xlim=[-10, 10],
    ylim=[-10, 10],
)

p1_2 = plot(
    fun_1_2,
    (x, -5, 3),
    legend=True,
    line_color="orange",
    show=False,
)

p1_3 = plot(
    fun_1_3,
    (x, 4, 10),
    legend=True,
    line_color="green",
    show=False,
)

p1_1.append(p1_2[0])
p1_1.append(p1_3[0])
p1_1.show()
```

### Zadanie 2

Legenda nie jest wyświetlana, gdyż przez nią wyrzuca błąd


```python
fun2 = sp.Piecewise(
    ((x - 5) ** 2 - 25, (x >= -5) & (x < 0)),
    (x**3 + 0.5, (x >= 0) & (x < 3)),
    (sp.Abs(sp.sin(3 * x) * 5 + 25), (x >= 3) & (x < 6)),
)

fun2_diff = sp.diff(fun2, x)

p2_1 = plot(
    fun2,
    xlabel="x",
    ylabel="y",
    title="Zadanie 2 - funkcja klamerkowa i jej pochodna (legenda)",
    line_color="blue",
    show=False,
    xlim=[-10, 10],
    ylim=[-10, 10],
)

p2_2 = plot(fun2_diff, line_color="red", show=False)

p2_1.append(p2_2[0])
p2_1.show()
```
