```python
import sympy as sp

sp.init_printing()
```

```python
t = sp.symbols("t")

r = 8  # m
w = 0.5  # rad/s
v = 0.8  # m/s
```

(a) Po jakim czasie aktor opuści scenę?

**Odp: 10s**


```python
czas_zejscia = r / v
czas_zejscia
```

(b) Ile razy obróci się scena aż aktor ją opuści?

**Odp: 2.5/π, czyli 0.796 obrotu**


```python
obroty_na_sekunde = w / (2 * sp.pi)
ilosc_obrotow = obroty_na_sekunde * czas_zejscia
print(ilosc_obrotow.evalf(10))
ilosc_obrotow
```

(c) Jaką drogę przebędzie aktor? Oblicz długość toru ruchu aktora po 5 i 10 sekundach.

**Odp: 6.7m po 5 sekundach i 22.25m po 10 sekundach**


```python
x_t = v * t * sp.cos(w * t)
y_t = v * t * sp.sin(w * t)

# pochodne pierwszego stopnia dla x_t i y_t
x_t_d1 = sp.diff(x_t, t)
y_t_d1 = sp.diff(y_t, t)

# wzór na długość łuku krzywej parametrycznej
wzor = sp.sqrt(x_t_d1**2 + y_t_d1**2)

dlugosc_luku_krzywej = sp.Integral(wzor)
dlugosc_luku_krzywej
```

Długość toru po 5 sekndach


```python
dlugosc_po_5_sek = sp.Integral(wzor, (t, 0, 5))
dlugosc_po_5_sek.n(10)
```

Długość toru po 10 sekundach


```python
dlugosc_po_10_sek = sp.Integral(wzor, (t, 0, 10))
dlugosc_po_10_sek.n(10)
```

(e) Wykonaj odpowiedni rysunek.


```python
wykres_droga = sp.plot_parametric(
    (x_t, y_t),
    (t, 0, czas_zejscia),
    xlim=[-r - 1, r + 1],
    ylim=[-r - 1, r + 1],
    show=False,
    size=(4, 4),
)
wykres_scena = sp.plot_parametric(
    (sp.cos(t) * r, sp.sin(t) * r),
    (t, 0, 2 * sp.pi),
    line_color="red",
    show=False,
    size=(4, 4),
)
wykres_droga.append(wykres_scena[0])
wykres_droga.show()
```
